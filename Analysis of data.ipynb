#### Correlation analysis to explore relationships between variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
file_path = 'Survey data option B.csv'
df = pd.read_csv(file_path)

# Drop non-numeric columns (like 'Start Date') before calculating correlation
numeric_df = df.select_dtypes(include=['number'])

# Check if there are any numeric columns
if numeric_df.shape[1] == 0:
    print("No numeric columns found.")
else:
    # Check if there are any missing values in the numeric DataFrame
    if numeric_df.isnull().all().all():
        print("All values in the numeric DataFrame are missing.")
    else:
        # Calculate correlation matrix
        correlation_matrix = numeric_df.corr()

        # Check if the correlation matrix is empty or contains only NaN values
        if correlation_matrix.isnull().all().all():
            print("Correlation matrix is empty or contains only NaN values.")
        else:
            # Visualize the correlation matrix using a heatmap
            plt.figure(figsize=(22, 10))
            sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
            plt.title('Correlation Matrix')
            plt.show()
```

### T-tests/ANOVA for comparing means across groups.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import warnings

# Load the CSV file into a pandas DataFrame
file_path = 'Survey data option B.csv'
df = pd.read_csv(file_path)

# Assuming 'Gender', 'DistributionChannel', 'UserLanguage' are categorical variables
# and the rest are numerical variables

# Filter out rows with missing values in the numerical columns
numerical_columns = ['Duration (in seconds)', 'ES1', 'ES2', 'ES3', 'SR1', 'SR2', 'SR3',
                      'ETS1', 'ETS2', 'ETS3', 'SCR1', 'SCR2', 'SCR3', 'MKR1', 'MKR2', 'MKR3', 'SC0']
df = df.dropna(subset=numerical_columns)

# Define the categorical variables for grouping
categorical_variables = ['Gender', 'Educational Qualific', 'Working experience', 'Position']

# Perform ANOVA and generate boxplots for each numerical column against each categorical variable
for column in numerical_columns:
    plt.figure(figsize=(15, 8))
    
    for cat_var in categorical_variables:
        plt.subplot(1, len(categorical_variables), categorical_variables.index(cat_var) + 1)
        
        # Check if each group has a sufficient number of data points
        valid_groups = [variable for variable in [cat_var] if len(df[variable].unique()) > 1]
        
        if not valid_groups:
            print(f"Insufficient unique categories in {cat_var}. Unable to perform ANOVA.")
        else:
            group_data = [df[df[variable] == category][column] for variable in valid_groups for category in df[variable].unique()]
            
            # Check if each group has more than one data point
            valid_group_data = [data for data in group_data if len(data) > 1]
            
            if not valid_group_data:
                print(f"Insufficient data points in {cat_var}. Unable to perform ANOVA.")
            else:
                try:
                    # Perform ANOVA
                    anova_result = f_oneway(*valid_group_data)
                    print(f"\nANOVA for {column} against {cat_var}:")
                    print(f"F-Statistic: {anova_result.statistic}, p-value: {anova_result.pvalue}")
                    
                    # Generate boxplot
                    sns.boxplot(x=cat_var, y=column, data=df)
                    plt.title(f'Boxplot of {column} across {cat_var}')
                except warnings.WarningMessage as e:
                    print(f"Warning: {e}")
    
    plt.tight_layout()
    plt.show()
```

### Regression analysis for predicting a dependent variable from one or more independent variables.

```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
file_path = 'Survey data option B.csv'  # Replace with the actual file path
df = pd.read_csv(file_path)

# Select relevant columns
dependent_variable = 'Duration (in seconds)'
independent_variables = ['ES1', 'ES2', 'ES3', 'SR1', 'SR2', 'SR3',
                          'ETS1', 'ETS2', 'ETS3', 'SCR1', 'SCR2', 'SCR3', 'MKR1', 'MKR2', 'MKR3', 'SC0']

# Filter out rows with missing values in the selected columns
selected_columns = [dependent_variable] + independent_variables
df = df.dropna(subset=selected_columns)

# Create a design matrix by adding a constant term to the independent variables
X = sm.add_constant(df[independent_variables])

# Fit the multiple linear regression model
model = sm.OLS(df[dependent_variable], X).fit()

# Print the summary of the regression
print(model.summary())

# Plot the regression lines for all independent variables
plt.figure(figsize=(15, 10))
for variable in independent_variables:
    plt.subplot(4, 4, independent_variables.index(variable) + 1)
    sns.scatterplot(x=df[variable], y=df[dependent_variable], label='Actual Data')
    sns.lineplot(x=df[variable], y=model.predict(X), color='red', label='Regression Line')
    plt.xlabel(variable)
    plt.ylabel('Duration (in seconds)')
    plt.title(f'Regression Analysis for {variable}')

plt.tight_layout()
plt.show()
```

#### Chi-square tests for associations between two categoricalvariables.

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the CSV file into a pandas DataFrame
file_path = 'Survey data option B.csv'  # Replace with the actual file path
df = pd.read_csv(file_path)

# Select the two categorical variables of interest
categorical_var1 = 'Gender'
categorical_var2 = 'Educational Qualific'

# Create a contingency table
contingency_table = pd.crosstab(df[categorical_var1], df[categorical_var2])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the results based on the p-value
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. There is a significant association.")
else:
    print("Fail to reject the null hypothesis. No significant association.")
```
